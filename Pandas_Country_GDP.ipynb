# seaborn

```python
import matplotlib.pyplot as plt # visualization
import seaborn as sns   # for advance visualizations  -> distribution plot, line plot
```

```python
plt.rcParams['figure.figsize']=9,5
import warnings
warnings.filterwarnings('ignore')   # OS Errors
```

```python
stats.head()
```

```python
stats['InternetUsers']
```

```python
# Distributions
plt.rcParams['figure.figsize']=9,5
vis1 = sns.distplot(stats["InternetUsers"])   # Univarient Analysis is Plot the graph using 1 variable is called Univarients
```

```python
vis1 = sns.distplot(stats["InternetUsers"], bins=10)
```

```python
# Box Plots

vis2 = sns.boxplot(data = stats, x='IncomeGroup', y='BirthRate')  # Bi Varient Analysis


# stright line is the Average
# stright line below is Lower Quartile
# stright line above is Upper Quartile
```

```python
# Outliers - Outliers in stats datapoints  which are far from other observation other group
# Outliers also called as Anomaly detection
# Outliers will inpact on Machine Learning Algortihm
# We adjust Outliers by implementing Probability function == logistic regression algorithm
```

```python
vis4 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate')
```

```python
vis3 = sns.lmplot(data=stats, x='InternetUsers', y='BirthRate', fit_reg=True)  # lm - linear model
```

```python
vis31 = sns.lmplot(data = stats,x = 'InternetUsers', y = 'BirthRate', fit_reg = False) #lm - linear model
```

```python
vis5 = sns.lmplot(data = stats,x = 'InternetUsers', y = 'BirthRate', fit_reg = False,hue = 'IncomeGroup') #hue - parameter for color
```
